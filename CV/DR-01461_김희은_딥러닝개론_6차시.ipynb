### Q4. 원-핫 인코딩이 무엇이고 왜 사용하는지 서술하시오.

원-핫 인코딩: 범주를 이진 벡터[0,1]로 변환하는 방식입니다.

이유: 범주 간 순서, 크기에 의미를 부여하지 않고 관계를 잘못 해석하는 것을 방지합니다.

### Q5. Sofmax가 무엇인지 정의하고, 어떠한 상황에서 주로 사용되는지 설명하시오.

Softmax: 주어진 입력을 각 요소의 확률 값으로 반환합니다. 반환값은 [0,1]사이의 값이고 반환값들의 합은 항상 1입니다.

### Q6. 아래 제공된 벡터 x가 softmax를 통과하면 어떻게 변하는지 구하시오.

$$

    x = [1,2,3,4]

$$

공식.png



```python
import numpy as np
x=[1,2,3,4]
def softmax(x):
    e_x = np.exp(x - np.max(x))
    return e_x / e_x.sum()

print(softmax(x))
```
