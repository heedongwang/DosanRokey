### Q4. 다음은 PyTorch를 사용하여 자동 미분을 수행하려는 코드이다. 그러나 코드에는 오류가 있다. 해당 오류를 수정하여 올바르게 자동 미분을 수행하라.


```python
import torch

x = torch.tensor([1, 2, 3], requires_grad=True ,dtype=torch.float64)
y = torch.tensor([4, 5, 6])

z = torch.sum(x * y)

loss = z

loss.backward()
x.grad
```

x에 requires_grad=True를 활성화 해줘야 하고, 데이터 타입을 float으로 변경해야합니다.


### Q5. torch의 zero_grad 함수는 무엇이며, 어떠한 역할을 하는가? 그리고 만약 zero_grad를 호출하지 않으면 어떤 문제가 발생하는가?

 zero_grad: 이전 기울기를 0으로 초기화합니다. 자동 미분을 사용 할때 backward 호출될 때 마다 값에 더해집니다.

 호출하지 않으면 이전단계의 기울기가 계속 새로운 기울기에 더해지며 영향을 줘 올바른 기울기 계산이 어렵습니다.
